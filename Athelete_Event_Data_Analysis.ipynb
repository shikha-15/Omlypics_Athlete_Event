```python
# athletes participation evolvment:
ath_number=df.groupby(['Year'])['ID'].nunique()

# bar plot for the athletes participation number:
ath_number.plot(kind='bar', figsize=(10,3))
plt.title('Atheletes Number By Season',color='red')
plt.xlabel('Year',color='green')
plt.ylabel('Atheltes Number',color='green')
plt.show()

```

```python
 # line plot for the athletes participation number:
 ath_number.plot(kind='line', figsize=(10,3))
 plt.title('Atheletes Number By Season',color='red')
 plt.xlabel('Year',color='green')
 plt.ylabel('Atheltes Number',color='green')
 plt.show()
```

```python
 # atheletes highest participationn season,city:
 highest_number=df.groupby(['Year','Season','City'])['ID'].nunique().sort_values(ascending=False).head(1)
 highest_number

```

```python

 # bar plot for the 10 first countries:
 country_most_medals=df.groupby(['NOC'])['Medal'].count().sort_values(ascending=False).head(10)
 country_most_medals.plot(kind='bar', figsize=(5,3))
 plt.title('Country wise Medals',color='red')
 plt.xlabel('Country',color='green')
 plt.ylabel('Medals',color='green')
 plt.show()
```

```python
# most gold medals, country, season: the 5 first countries:
country_most_gold_medals=df[df['Medal']=='Gold'].groupby(['Year','NOC','Season'])['Medal'].count().sort_values(ascending=False).head(5)
print(country_most_gold_medals)
```
